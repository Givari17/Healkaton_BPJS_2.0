### **Waktu Kunjungan**
Jenis variabel:

    PNK03 => Tanggal kunjungan Tanggal melakukan kunjungan

    PNK04 => Tanggal tindakan Tanggal melakukan tindakan

    PNK05 => Tanggal pulang Tanggal menyelesaikan kunjungan

Untuk Variabel 'Lama Kunjungan' dan 'Lama Tindakan' merupakan jenis data ordinal dengan urutan sebagai berikut:

Lama Kunjungan

    0  -->  Menunjukan lama Kunjungan pasien hanya satu hari

    1  -->  Menunjukan lama Kunjungan pasien hanya dua hari

    2  -->  Menunjukan lama Kunjungan pasien hanya tiga hari
    
    dst...

```python
# Melihat waktu kunjungan berdasarkan bulan.
# Menambahkan kolom 'bulan'
df['bulan'] = df['PNK03'].dt.month

# Membuat kamus pemetaan bulan
kamus_bulan = {1: 'Januari', 2: 'Februari', 3: 'Maret', 4: 'April', 5: 'Mei',
               6: 'Juni', 7: 'Juli', 8: 'Agustus', 9: 'September', 10: 'Oktober',
               11: 'November', 12: 'Desember'}

# Mengubah isi kolom 'bulan' menjadi nama bulan
df['bulan'] = df['bulan'].map(kamus_bulan)

df.head()
```

```python
sns.catplot(data=df, x="bulan", kind="count", palette="ch:.25")
```

```python
#PNK12 Vriabel Segmen peserta saat mengakses FKTP
df['PNK12'].unique()
```

```python
# waktu Lama Kunjungan Pasien
df['Lama_Kunjungan'] = (df['PNK05'].dt.day - df['PNK03'].dt.day).abs()
df['Lama_Kunjungan']

#mengkodekan kelas lama kunjungan
for i in range(len(df)):
  if df['Lama_Kunjungan'][i] <= 1:
    df['Lama_Kunjungan'][i] = 2
  elif df['Lama_Kunjungan'][i] >= 10:
    df['Lama_Kunjungan'][i] = 7
  elif df['Lama_Kunjungan'][i] == 0:
    df['Lama_Kunjungan'][i] = 1
  else:
    df['Lama_Kunjungan'][i] = 3
```

# **Asosiation Rule Mining Modell**

```python
df['PNK16'].unique()
```

```python
# Diagnosis Pada Kepesertaan FKTP NON KAPITASI Promotif

Diagnosa_Prom = (df[df['PNK11'] == 'PROMOTIF']
                 .groupby(['PSTV02','PNK16'])['PNK17'].sum().unstack().reset_index().fillna(0)
                 .set_index('PSTV02'))


Diagnosa_RJTP = (df[df['PNK11'] == 'RJTP']
                 .groupby(['PSTV02','PNK16'])['PNK17'].sum().unstack().reset_index().fillna(0)
                 .set_index('PSTV02'))

Diagnosa_RITP = (df[df['PNK11'] == 'RITP']
                 .groupby(['PSTV02','PNK16'])['PNK17'].sum().unstack().reset_index().fillna(0)
                 .set_index('PSTV02'))
```

```python
Diagnosa_Prom
```

```python
Diagnosa_Prom.info()
```

```python
#Metransformasikan menjadi biner
def hot_encode(x):
  if(x<=0):
    return False
  if(x>=1):
    return True
```

```python
df_encode = Diagnosa_Prom.applymap(hot_encode)
Diagnosa_Prom  = df_encode

df_encode = Diagnosa_RJTP.applymap(hot_encode)
Diagnosa_RJTP  = df_encode

df_encode = Diagnosa_RITP.applymap(hot_encode)
Diagnosa_RITP  = df_encode

Diagnosa_Prom
```

```python
#memanggil modul ARm
from mlxtend.frequent_patterns import apriori, association_rules

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
import warnings
warnings.filterwarnings('ignore')
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#Membangun model ARM untuk Tindakan kepesertaan PROMOTIF
frq_items = apriori(Diagnosa_Prom,min_support=0.05, use_colnames=True)

#Mengumpulkan aturan yang disimpulkan dalam dataframe
rules = association_rules(frq_items, metric="lift", min_threshold=1)
rules = rules.sort_values(['confidence','lift'], ascending=[False, False])
rules.head()
```

```python
#Membangun model ARM untuk Tindakan kepesertaan RJTP
frq_items = apriori(Diagnosa_RJTP,min_support=0.005, use_colnames=True)

#Mengumpulkan aturan yang disimpulkan dalam dataframe
rules = association_rules(frq_items, metric="lift", min_threshold=1)
rules = rules.sort_values(['confidence','lift'], ascending=[False, False])
rules.head()
```
